```python
### dividing the age into 4 age groups 
import numpy as np
cancer["age_group"] = np.where((cancer["age"] > 0) & (cancer["age"] <= 15),"Children",
                        np.where((cancer["age"] >= 16) & (cancer["age"] <= 25),"Teenegars",
                        np.where((cancer["age"] >25 ) & (cancer["age"] <= 59),"Adult","Old")))
cancer
```

```python
### checking how many patient across all age group
cancer["age_group"].value_counts()
```

```python
### using bmi to categorise all patient weight 
cancer["weight"] = np.where((cancer["bmi"] > 0) & (cancer["bmi"] < 18.5),"Underweight",
                        np.where((cancer["bmi"] >= 18.5) & (cancer["bmi"] <= 24.9),"Normal weight",
                        np.where((cancer["bmi"] > 24.9 ) & (cancer["bmi"] <= 29.9),"Overweight","Obese")))
cancer
```

```python
### categorising cholesteral level of all patients 
cancer["Cholesterol_status"] = np.where((cancer["cholesterol_level"] > 0) & (cancer["cholesterol_level"] < 200),"low",
                        np.where((cancer["cholesterol_level"] >= 200) & (cancer["cholesterol_level"] <= 239),"medium",
                        np.where((cancer["cholesterol_level"] >=240) & (cancer["cholesterol_level"] <=400),"high","")))
cancer
```

```python
### converting the diagnosis_date from object to date and time fomart 
cancer["diagnosis_date"] = cancer["diagnosis_date"].astype("datetime64[ns]")
cancer
```

```python
### converting the end_treatment_date t0 date and time fomart
cancer["end_treatment_date"] = cancer["end_treatment_date"].astype("datetime64[ns]")
cancer
```

```python
### getting the number of days of treatment for all patients 
cancer["NUM_OF_DAYS_TREATMENT"] = cancer["end_treatment_date"] - cancer["diagnosis_date"]
cancer
```

```python
### survival rate across all cancer stages
cvd=cancer.groupby(["cancer_stage"])["survived"].value_counts().reset_index()
cvd
```

```python
# Assuming your value_counts column is named 0 after reset_index
cvd.columns = ["cancer_stage", "survived", "count"]
```

```python
# Pivot the data for plotting
stage = cvd.pivot(index="cancer_stage", columns="survived", values="count").fillna(0)
stage
```

```python
# Plot as a grouped bar chart
stage.plot(kind="bar", figsize=(6, 3))

plt.title("Survival Counts by Cancer Stage")
plt.xlabel("Cancer Stage")
plt.ylabel("Count")
plt.legend(title="Survived", labels=["No (0)", "Yes (1)"])
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
